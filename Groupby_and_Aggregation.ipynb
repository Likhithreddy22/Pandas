```python
import pandas as pd
```

```python
data = {
    'Department':['Sales','Sales','IT','IT','HR','HR'],
    'Employee':['Alice','Bob','Charlie','David','Eve','Frank'],
    'Salary':[50000,60000,70000,80000,90000,45000],
    'Experience':[3,5,3,8,6,1]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
grouped = df.groupby('Department')
```

```python
for name,group in grouped:
    print(name)
    print(group)
```

```python
avg_sal = grouped['Salary'].mean()
```

```python
avg_sal
```

```python
sal_state = df.groupby('Department')['Salary'].agg(['mean','min','max','std'])
```

```python
sal_state
```
