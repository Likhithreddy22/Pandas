```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1,2],'B':[3,4]})
```

```python
df2 = pd.DataFrame({'A':[5,6],'B':[7,8]})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
df3 = pd.DataFrame({'id':[1,2,3],'name':['Alice','Bob','Charlie']})
```

```python
df4 = pd.DataFrame({'id':[2,3,4],'score':[85,92,78]})
```

```python
pd.merge(df3,df4,on = 'id')
```

```python
pd.merge(df3,df4,on = 'id',how = 'left')
```

```python
pd.merge(df3,df4,on = 'id',how = 'right')
```
