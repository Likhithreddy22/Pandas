```python
df['math_group'] = pd.cut(df['Maths'],bins=[20,60,100],labels=['20-60','60-100'])
```

```python
df['name_length'] = df['Student_Name'].apply(len)
```

```python
import numpy as np
df['math_grade'] = np.where(df['Maths']>70,'A','B')
```

```python
conditions = [ df['Chemistry']<40,
               df['Chemistry'].between(40,75),
               df['Chemistry']>75
             ]
choices = ['Low','Medium','High']
```

```python
df['Chemistry_category'] = np.select(conditions,choices,default = 'Unknown')
```
