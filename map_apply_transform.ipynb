```python
import pandas as pd
```

```python
df = pd.DataFrame({'name':['Alice','Bob','Charlie'],
                   'gender':['F','M','M']})
```

```python
df['gender'].map({'F':'Female','M':'Male'})
```

```python
df['name'].map(lambda x:x.upper())
```

```python
df['name'].apply(len)
```

```python
df = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
```

```python
df.apply(lambda row : row['A']+row['B'],axis = 1)
```

```python
df.apply(lambda col:col.mean())
```

```python
df = pd.DataFrame({'group':['A','A','B','B'],'value':[1,2,3,4]})
```

```python
df.groupby('group')['value'].transform(lambda x:x/x.sum())
```

```python
df.groupby('group')['value'].transform(['mean','sum'])
```
