```python
import numpy as np
```

```python
import pandas as pd
```

```python
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', 
             '2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Region': ['North', 'North', 'North', 'North', 
               'South', 'South', 'South', 'South'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 120, 130, 180, 220, 140],
    'Quantity': [10, 15, 20, 12, 13, 18, 22, 14]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
pivot1 = pd.pivot_table(df,values = 'Sales',index = 'Region',columns = 'Product',aggfunc='sum')
```

```python
pivot1
```

```python
pivot2 = pd.pivot_table(df,values = 'Sales' , index = ['Region','Product'] , columns = 'Date' , aggfunc = 'sum')
```

```python
pivot2
```

```python
pivot3 = pd.pivot_table(df, values= ['Sales','Quantity'], index = 'Region',columns = 'Product',aggfunc = 'sum')
```

```python
pivot3
```

```python
df_missing = df.copy()
```

```python
df_missing.loc[0,'Sales'] = np.nan
```

```python
df_missing
```

```python
pivot_with_nan = pd.pivot_table(df_missing,values = 'Sales', index ='Region' , columns = 'Product' , aggfunc = 'mean')
```

```python
pivot_with_nan
```

```python
pivot_filled = pd.pivot_table(df_missing,values = 'Sales',index = 'Region',columns = 'Product',aggfunc = 'mean',fill_value = 0)
```

```python
pivot_filled
```

```python
pivot_dropna = pd.pivot_table(df_missing,values = 'Sales',index = 'Region',columns = 'Product',aggfunc = 'mean',dropna = True)
```

```python
pivot_dropna
```

```python
pivot_margin = pd.pivot_table(df,values = 'Sales',index = 'Region' , columns = 'Product', aggfunc = 'sum', margins = True, margins_name = 'Total')
```

```python
pivot_margin
```
